```python
df = pd.DataFrame({
    "이름": ["Evan", "Sara", "Chris", "Messi", "Woods", "Jordan"], 
    "국어": [70, 90, None, None, 100, 85], 
    "영어": [65, 100, None, 90, 70, 80]
})
df 
```

## 행추가 

```python
df.loc[6, "이름":"영어"] = ["Kim", 50, 60]
df
```

```python
%%timeit

def calculate_grade(score):
    if score >= 90:
        return 'A'
    elif score >= 80:
        return 'B'
    elif score >= 70:
        return 'C'
    else:
        return 'F'
df['등급'] = df['국어'].apply(lambda x: calculate_grade(x))
```

```python
%%timeit 

df['등급'] = df['국어'].apply(lambda score: 'A' if score >= 90 else ('B' if score >= 80 else ('C' if score >= 70 else 'F')))
df
```

```python
%%timeit 

df['등급'] = ''

for i in range(0, df.shape[0]):
    if df.loc[i, '국어'] >= 90:
        df.loc[i, '등급'] = "A"
    elif df.loc[i, '국어'] >= 80:
        df.loc[i, '등급'] = "B"
    elif df.loc[i, '국어'] >= 70:
        df.loc[i, '등급'] = "C"
    else:
        df.loc[i, '등급'] = "F"
df
```

# 국어점수가 90점 이상이면 "A"
# 국어점수가 80점~89점 "B"
# 국어점수가 70점~79점 "C"
# 나머지는 F

- numpy.where, 조건식 한번만 적용
if 조건식, 참일 때, 거짓일 때

```python
# np.where(조건식, 참, np.where(조건식, 참, (np.where(조건식, 참))))
# np.select(다중조건식, 다중결괏값, default==> else 조건)
```

```python
%%timeit

condition_list = [(df["국어"] >= 90), (df['국어'] >= 80) & (df['국어'] < 90), (df['국어'] >= 70) & (df['국어'] < 80)]
choice_list    = ["A",                 "B",                                    "C"]
df['점수'] = np.select(condition_list, choice_list, default = "F")
df
```
