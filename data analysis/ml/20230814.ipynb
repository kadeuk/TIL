```python
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/ethanweed/pythonbook/main/Data/harpo.csv")
df
```

```python
df.groupby('tutor')['grade'].agg(['mean', 'std', 'count']).reset_index()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 데이터프레임 그룹화 및 집계
agg_df = df.groupby('tutor')['grade'].agg(['mean', 'std', 'count']).reset_index()

# 그래프 그리기
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(14,6))

# Anastasia의 데이터만 선택하여 첫 번째 그래프에 그립니다.
anastasia_data = agg_df[agg_df['tutor'] == 'Anastasia']
axes[0].bar(anastasia_data['tutor'], anastasia_data['mean'], yerr=anastasia_data['std'], capsize=10)
axes[0].set_xlim(50,100)
axes[0].set_ylim(0,7)
axes[0].set_title('Anastasia')

# Bernadette의 데이터만 선택하여 두 번째 그래프에 그립니다.
bernadette_data = agg_df[agg_df['tutor'] == 'Bernadette']
axes[1].bar(bernadette_data['tutor'], bernadette_data['mean'], yerr=bernadette_data['std'], capsize=10)
axes[1].set_xlim(50,100)
axes[1].set_ylim(0,7)
axes[1].set_title('Bernadette')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 데이터 준비
data = {
    'tutor': ['Anastasia', 'Bernadette'],
    'mean': [74.533333, 69.055556],
    'std': [8.998942, 5.774918],
    'count': [15, 18]
}
agg_df = pd.DataFrame(data)

# 그래프 그리기
fig, ax = plt.subplots(figsize=(8, 5))

# 막대 그래프 그리기
ax.bar(agg_df['tutor'], agg_df['mean'], yerr=agg_df['std'], color=['blue', 'green'], alpha=0.7)
ax.set_ylim(0, 100)
ax.set_title('Mean Grades by Tutor with Std Deviation')
ax.set_ylabel('Mean Grade')

plt.tight_layout()
plt.show()
```
